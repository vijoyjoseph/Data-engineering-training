```python

import pandas as pd

class ECommerceData:
    def __init__(self, data_frame):
        self.df = data_frame
        self.df['Total Sales'] = self.df['Product Price'] * self.df['Quantity Sold']
        self.df['Profit Margin'] = self.df['Total Sales'] - (self.df['Product Price'] * self.df['Quantity Sold'] * 0.2)  # Example calculation for cost

    def get_sales_by_category(self):
        return self.df.groupby('Product Category')['Total Sales'].sum().sort_values(ascending=False)

    def get_top_products(self, top_n=5):
        return self.df.groupby('Product ID')['Total Sales'].sum().nlargest(top_n)

    def get_least_products(self, bottom_n=5):
        return self.df.groupby('Product ID')['Total Sales'].sum().nsmallest(bottom_n)

    def calculate_profit_margin(self):
        return self.df.groupby('Product ID')['Profit Margin'].mean().sort_values(ascending=False)

 
#df = pd.read_csv('transactions.csv')  # Load your dataset
ecommerce_data = ECommerceData(df)

 
print(ecommerce_data.get_sales_by_category())

 
print(ecommerce_data.get_top_products())
 
print(ecommerce_data.get_least_products())
 
print(ecommerce_data.calculate_profit_margin())
```
